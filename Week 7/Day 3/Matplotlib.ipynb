```python
import matplotlib.pyplot as plt
import pandas as pd
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
df
```

```python
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']
virginica = df[df['species'] == 'virginica']
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
```

```python
plt.bar(df['species'], df['species'].count())
```

```python
plt.scatter(setosa['petal_length'], setosa['petal_width'], c='red')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
plt.scatter(versicolor['petal_length'], versicolor['petal_width'], c='black')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
plt.scatter(virginica['petal_length'], virginica['petal_width'], c='green')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
plt.boxplot([setosa['sepal_length'], versicolor ['sepal_length'], virginica['sepal_length']])
plt.xticks([1, 2, 3], ['setosa', 'versicolor', 'virginica'])
plt.ylabel('Sepal Length')
```

```python
plt.hist(df['sepal_width'], bins = 10)
plt.xlabel('Sepal Width')
plt.ylabel('Count')
```

```python
plt.scatter(setosa['sepal_length'], setosa['petal_length'], c='blue')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

```python
plt.scatter(versicolor['sepal_length'], versicolor['petal_length'], c='pink')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

```python
plt.scatter(virginica['sepal_length'], virginica['petal_length'], c='gold')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

```python
df.groupby('species')['petal_length'].mean()
plt.bar(df.groupby('species')['petal_length'].mean().index, df.groupby('species')['petal_length'].mean().values)
plt.xlabel('Species')
plt.ylabel('Mean Petal Length')
```
