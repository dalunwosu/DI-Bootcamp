```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
arr1 = np.random.randint(0,75,100)
arr1
```

```python
arr1 = arr1.reshape((50,2))
arr1
```

```python
plt.scatter(arr1[:, 0], arr1[:, 1])
plt.xlabel('Column 1')
plt.ylabel('Column 2')
```

```python
plt.hist(arr1[:, 0])
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.title('Column 1')
```

```python
plt.hist(arr1[:, 1],color = 'red')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.title('Column 2')
```


- Make a histogram of the Fare feature, colored green.


- Produce a similar plot to the pie charts above except instead of splitting the data by gender, split it by Pclass.


- Can you make any observations about these plots that visually correlate the Fare by Survived distribution?
- Save that figure as a png file.



```python
tt = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
tt
```

```python
tt['Pclass'].unique()
```

```python
first = tt[tt.Pclass== 1].Survived.value_counts(normalize = True)
second = tt[tt.Pclass== 2].Survived.value_counts(normalize = True)
third = tt[tt.Pclass== 3].Survived.value_counts(normalize = True)
```

```python
fig, ax = plt.subplots(1,3,figsize = (13,6))
ax[0].pie(first,labels = first.index.map({0: 'Did Not Survive',1:'Survived'}),autopct = '%1.1f%%',)
ax[0].set_title('First Class Passengers')
ax[1].pie(second,labels = second.index.map({0: 'Did Not Survive',1:'Survived'}),autopct = '%1.1f%%',)
ax[1].set_title('Second Class Passengers')
ax[2].pie(third,labels = third.index.map({0: 'Did Not Survive',1:'Survived'}),autopct = '%1.1f%%',)
ax[2].set_title('Third Class Passengers')
```
