```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep = "|")
users
```

```python
users.groupby("occupation").age.mean()
```

```python
users.rename(columns={"sex":"gender"}, inplace=True)
```

```python
users
```

```python
(users.groupby("occupation")["gender"].value_counts(normalize=True)*100)
```

```python
gendersn = pd.DataFrame((users.groupby("occupation")["gender"].value_counts(normalize=True)*100))
gendersn.rename(columns={"gender":"percent"}, inplace=True)
gendersn
```

```python
new_gen = pd.DataFrame(gendersn.reset_index(level=["occupation", "gender"]))
new_gen
```

```python
males = new_gen[new_gen["gender"] == "M"].sort_values(by="percent",ascending=False)
males
```

```python
users.groupby("occupation").age.agg([min, max])
```
